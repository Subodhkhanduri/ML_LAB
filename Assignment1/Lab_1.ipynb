```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("supermarket_sales - Sheet1.csv")
```

```python
df.head()
```

```python
df.Date = pd.to_datetime(df['Date'])
```

```python
df.head()
```

```python
df_Sort_date = df.sort_values(by='Date')
```

```python
df_Sort_date
```

```python
x = df_Sort_date.iloc[:, 10].values
y = df_Sort_date.iloc[:, 9].values
```

```python
df['Date'] = pd.to_datetime(df['Date'])
x = (df['Date'] - pd.Timestamp('1970-01-01')).dt.days.values.reshape(-1, 1)
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x, y)
```

```python
plt.scatter(x, y, color = 'red')
plt.plot(x, regressor.predict(x), color = 'blue')
plt.title('Total sales vs Time')
plt.xlabel('Time')
plt.ylabel('Sales')
plt.show()
```
