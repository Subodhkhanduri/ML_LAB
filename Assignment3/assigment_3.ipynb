```python
numeric_col = df.select_dtypes(include=['int64','float64']).columns
```

```python
numeric_col
```

```python
def plot_numerical_col(df,column):
    plt.figure(figsize=(8,6))
    sns.boxplot(x= df[column], palette='Set3')
    plt.title(f'Distribution of {column}')
    plt.show()

for col in numeric_col:
    plot_numerical_col(df,col)
```

```python
cat_col = df.select_dtypes(include='object').columns
```

```python
def categrocial_column(df,column):
    plt.figure(figsize=(8,6))
    sns.countplot(data=df, x=column, palette='Set3', edgecolor='black')
    plt.title(f'Distribution of {column}', fontweight='bold')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

for col_name in cat_col:
    categrocial_column(df,col_name)
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()

for column in cat_col:
    df[column] = encoder.fit_transform(df[column])
```

```python
df.head()
```

```python
correlation_matrix = df.corr()
correlation_target = correlation_matrix['Accident'].sort_values(ascending=False).round(3)
print(correlation_target)
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(),annot=True, cmap='coolwarm', fmt='.3f', linewidth=0.5)
plt.show()
```

```python
threshold = 0.02

correlation_matrix = df.corr()
high_corr_features = correlation_matrix.index[abs(correlation_matrix['Accident'])>threshold].tolist()
high_corr_features.remove('Accident')
print(high_corr_features)

x = df[high_corr_features]
y = df['Accident']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
print("Training Test size:", x_train.shape)
print("Testing Test Size:", x_test.shape)
```

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
dt.fit(x_train,y_train)
```

```python
y_pred = dt.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
Accuracy = accuracy_score(y_test, y_pred)
confusion = confusion_matrix(y_test, y_pred)
classification = classification_report(y_test, y_pred)
```

```python
print(f"Accuray Score:{Accuracy:.4f}")
print(f"Confusion Matrix:\n {confusion}")
print(f"Classification Report:\n {classification}")
```
